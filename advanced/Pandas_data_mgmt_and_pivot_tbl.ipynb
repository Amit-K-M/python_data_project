```python
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset= load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_original = df.copy()
df_alt = df_original.copy()
```

```python
median_salary = df_alt['salary_year_avg'].median()

df_alt['salary_year_avg'] = df_alt['salary_year_avg'].fillna(median_salary)

df_alt.loc[:5,'salary_year_avg']

```

```python
df_original.loc[:5,'salary_year_avg']
```

```python
df.sample(n=5)
```

```python
df.sample(frac=0.1,replace=False)
```

```python
df.groupby(['job_country','job_title_short'])['salary_year_avg'].agg(['median','max','min']).dropna()
```

```python
df_job_country_salary = df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst','Data Engineer','Data Scientist']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar')
plt.xlabel('')
plt.ylabel('Median Salary in USD$')
plt.title('Median Salary by Job Title and Country')
plt.xticks(rotation = 45, ha='right')
plt.show()
```

```python
df_ind = df[df['job_country'] == 'India'].copy()

df_ind['job_posted_month'] = df_ind['job_posted_date'].dt.strftime('%B')

df_ind_pivot = df_ind.pivot_table(index='job_posted_month' , columns= 'job_title_short', aggfunc='size')

df_ind_pivot
```

```python
df_ind_pivot = df_ind_pivot.reset_index()

df_ind_pivot
```

```python
df_ind_pivot['job_posted_month_no'] = pd.to_datetime(df_ind_pivot['job_posted_month'], format='%B').dt.month

df_ind_pivot = df_ind_pivot.sort_values('job_posted_month_no')

df_ind_pivot = df_ind_pivot.set_index('job_posted_month')
df_ind_pivot



```

```python
df_ind_pivot = df_ind_pivot.drop(columns= 'job_posted_month_no')

df_ind_pivot
```

```python
df_ind_pivot.plot(kind='line')
```

```python
top_3 = (df_ind['job_title_short']
         .value_counts()
         .head(3)
         .index.
         to_list())

top_3
```

```python
df_ind_pivot[top_3].plot(kind='line')
plt.title('Monthly job postings for top jobs in India')
plt.ylabel('')
plt.xlabel('2023')
plt.legend()
plt.show()
```

```python
df_ind2 = df[df['job_country'] == 'India'].copy()

df_ind2['job_posted_month'] = df_ind2['job_posted_date'].dt.to_period('M')

df_ind2_pivot = df_ind2.pivot_table(index='job_posted_month', columns='job_title_short',aggfunc='size')

df_ind2_pivot
```

```python
df_ind2_pivot[top_3].plot(kind='line')
plt.ylabel('')
plt.xlabel('2023')
plt.legend()
plt.title('Monthly job postings in India')
plt.show()
```
